```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=plt.imread("img/RGB-pic.jpg") 
plt.imshow(a)
```

```python

red=a.copy()
green=a.copy()
blue=a.copy()
red[:,:,1:3]=0
green[:,:,::2]=0
blue[:,:,:2]=0
fig,ax=plt.subplots(2,2)
fig.set_size_inches(15,15)
ax[0,0].imshow(a)
ax[0,1].imshow(red)
ax[1,0].imshow(green)
ax[1,1].imshow(blue)
```

```python
b=np.array([0.299,0.587,0.114])
x=np.dot(a,b)
plt.imshow(x,cmap="gray")
```
